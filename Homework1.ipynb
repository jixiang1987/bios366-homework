```python
def GA(pop_size, iterations, mutation_rate, Title, independence = None):
    
    aic = lambda g: g.nobs * np.log((g.resid**2).sum()/g.nobs) + 2*len(g.beta)
    aic_best = []
    best_solution = []
    aic_history = []
    
    # Initialize genotype
    current_gen = np.random.binomial(1, 0.5, pop_size*ncols).reshape((pop_size, ncols))
    
    def calculate_fitness(aic_values):
        P = len(aic_values)
        aic_rank = (-aic_values).argsort().argsort()+1.
        return 2.*aic_rank/(P*(P+1.))
    
    for i in range(iterations):
        
        # Get phenotype
        current_phe = [predictors[predictors.columns[g.astype(bool)]] for g in current_gen]
        # Calculate AIC
        current_aic = np.array([aic(pd.ols(y=logsalary, x=x)) for x in current_phe])
        # Get lowest AIC
        aic_best.append(current_aic[np.argmin(current_aic)])
        best_solution.append(current_gen[np.argmin(current_aic)])
        
        # Calculate fitness according to AIC rank
        fitness = calculate_fitness(current_aic)
        
        # Choose first parents according to fitness
        moms = np.random.choice(range(pop_size), size=pop_size/2, p=fitness)
        if independence:
            # Choose first parents according to fitness
            dads = np.random.choice(range(pop_size), size=pop_size/2, p=fitness)
        else:
            # Choose second parents randomly
            dads = np.random.choice(range(pop_size), size=pop_size/2, p=None)
        
        next_gen = []
        for x,y in zip(current_gen[moms], current_gen[dads]):
            # Crossover
            cross = np.random.randint(0, ncols)
            child1 = np.r_[x[:cross], y[cross:]]
            child2 = np.r_[y[:cross], x[cross:]]
            # Mutate
            m1 = np.random.binomial(1, mutation_rate, size=ncols).astype(bool)
            child1[m1] = abs(child1[m1]-1)
            m2 = np.random.binomial(1, mutation_rate, size=ncols)
            child2[m2] = abs(child1[m2]-1)
            next_gen += [child1, child2]
            
        # Increment generation
        current_gen = np.array(next_gen)
        # Store AIC values
        aic_history.append(current_aic)
    plt.figure()
    plt.title(Title)
    plt.plot(aic_best)
    plt.figure()
    plt.title(Title)
    for i,x in enumerate(aic_history):
        plot(np.ones(len(x))*i, -x, 'r.', alpha=0.3)
```

Use the combinatorial optimization method of your choice to obtain a solution to the traveling salesman problem for the Brazilian cities described in the lecture notes, using minimum total distance as the criterion. Use the the first city listed in the dataset as "home" (*i.e.* the trip must start and end there. I will award 5 bonus points to the best solution!

```python
%pylab inline
import pandas as pd
import matplotlib.pyplot as plt
import time
from pandas import *
from mpl_toolkits.basemap import Basemap

def parse_latlon(x):
    d, m, s = map(float, x.split(':'))
    ms = m/60. + s/3600.
    if d<0:
        return d - ms
    return d + ms

cities =  pd.read_csv('../data/brasil_capitals.txt', 
                      names=['city','lat','lon'])[['lat','lon']].applymap(parse_latlon)

def sum_distance(cities, run):
    s = 0
    for i in arange(25):
        x1 = -cities.values[run[i], 0]
        y1 = cities.values[run[i], 1]
        x2 = -cities.values[run[i+1], 0]
        y2 = cities.values[run[i+1], 1]
        s += sqrt((x1-x2)**2 + (y1-y2)**2)
    x1 = -cities.values[run[25], 0]
    y1 = cities.values[run[25], 1]
    x2 = -cities.values[run[0], 0]
    y2 = cities.values[run[0], 1]
    s += sqrt((x1-x2)**2 + (y1-y2)**2)
    return s

# Initial run
run = np.concatenate([[0], np.random.permutation(range(1, len(cities)))])

# Number of iterations
iterations = 2000

# Optimization
runs_distance = np.empty(iterations)
run_current = run

start_time = time.time()

for i in range(iterations):
    
    run_distance = sum_distance(cities, run_current)
    run_next = run_current
    
    for k in arange(25)[1:]:
        run_step = run_current.copy()
        kk = np.random.randint(k, 25, 1)    
        run_step[k], run_step[kk] = run_step[kk], run_step[k]    # City k switch once randomly with another one
        step_distance = sum_distance(cities, run_step)

        if step_distance < run_distance:  
            run_next = run_step.copy()
            run_distance = step_distance
        
    run_current = run_next.copy()
    runs_distance[i] = run_distance
        
    runs = run_current
 
print 'Optimized scheme is\n {0}'.format(runs)
end_time = time.time()
print("Elapsed time was %g seconds" % (end_time - start_time))
```

## Question 4

Suppose $y$ has a binomial distribution with parameters $n$ and $p$, and we are interested in the log-odds value $\theta = \log(p/(1 − p))$. Our prior for $\theta$ is that $\theta \sim N(\mu, \sigma^2)$. It follows that the posterior density of $\theta$ is given, up to a proportionality constant, by:

$$g(\theta | y) \propto \frac{\exp(y\theta)}{(1 + exp(\theta))^n} \exp\left[\frac{-(\theta − \mu)^2}{2\sigma^2}\right]$$

For example, suppose we are interested in learning about the probability that a possibly-biased coin lands heads when tossed. *A priori* we believe that the coin is fair, so we assign $\theta$ a $N(0,.25)$ prior. We toss the coin $n = 5$ times and obtain $y = 5$ heads.

1. Using a normal approximation to the posterior density, compute the probability that the coin is biased toward heads (i.e., that θ is posi- tive).
2. Using the prior density as a proposal density, design a rejection algo- rithm for sampling from the posterior distribution. Using simulated draws from your algorithm, approximate the probability that the coin is biased toward heads.
3. Using the prior density as a proposal density, simulate values from the posterior distribution using the SIR algorithm. Approximate the probability that the coin is biased toward heads.

```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fmin_bfgs
from scipy import stats
```

```python
# Posterior(Log-transformed)
def log_odds_post(theta, mu, sig, n, y):
    post = y*theta
    post -= n*np.log(1 + exp(theta))
    post += -(theta - mu)**2/(2*sig**2)
    return post

# Build min objective function
log_odds_post_min = lambda *args: -log_odds_post(*args)

# Give an initial guess
init_value = 1.0

# Optimization
opt = fmin_bfgs(log_odds_post_min, init_value, args = (0, .5, 5, 5), full_output = True)

# Get mean and var of parameters
mode, var = opt[0], opt[3]

# Output result
print '\nMode is {0}'.format(mode[0])
print 'Var is {0}'.format(var[0][0])

# Calculate CDF that theta is positive
Pr_theta_positive = 1 - stats.norm.cdf(0, loc = mode[0], scale = np.sqrt(var[0][0]))
print '\nProbability that theta is positive is {0}'.format(Pr_theta_positive)

# Plot positive area in PDF
x = np.linspace(-3, 3, 100)
y = stats.norm.pdf(x, loc = mode[0], scale = np.sqrt(var[0][0]))
plt.figure()
plt.plot(x, y)
plt.xlabel('$\theta$')
plt.ylabel('PDF')
plt.grid(True)
plt.fill_betweenx(y, x, where= x>0)
```

```python
#--------Find constant c--------
# log(f) - log(g)
def calc_diff(theta, mu, sig, n, y):
    return log_odds_post(theta, mu, sig, n, y) - stats.norm.logpdf(theta, loc = 0, scale = .25)

# Build min objective function
calc_diff_min = lambda *args: -calc_diff(*args)

# Give an initial guess
init_value = 1.0

# Optimization
opt = fmin_bfgs(calc_diff_min, init_value, args = (0, .5, 5, 5), full_output = True)

# Get constant c
c = opt[1]
print 'Constant c is {0}'.format(c)

#--------Rejection sampling algorithm--------
# Define rejection function
def reject(post, S, mu, n, data_n, data_y, c):
    
    # Draw samples from g(theta)
    theta = stats.norm.rvs(loc = mu, scale = S, size = n)
    
    # Calculate probability under g(theta)
    gvals = np.array([stats.norm.logpdf(t, loc = mu, scale = S) for t in theta])

    # Calculate probability under f(theta)
    fvals = np.array([post(t, mu, np.sqrt(S), data_n, data_y) for t in theta])
    
    # Calculate acceptance probability
    p = np.exp(fvals - gvals + c)
    return theta[np.random.random(n) < p]

# Sampling
nsamples = 100000
sample = reject(log_odds_post, .25, 0, nsamples, 5, 5, c)


# Calculate CDF that theta is positive
print '\nProbability that theta is positive is {0}'.format(np.mean(sample>0))

# Plot positive area in PDF
plt.figure()
x = np.linspace(-3, 3, 100)
mypdf = stats.gaussian_kde(sample)
plt.plot(x, mypdf(x))
plt.xlabel('$\theta$')
plt.ylabel('PDF')
plt.grid(True)
plt.fill_betweenx(mypdf(x), x, where= x>0)
```
